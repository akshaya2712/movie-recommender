### removing spaces from names

```python
def removespace(word):
    l=[]
    for i in word:
        l.append(i.replace(" ",""))
    return l
```

```python
movies['overview']=movies['overview'].apply(removespace)
movies['genres']=movies['genres'].apply(removespace)
movies['keywords']=movies['keywords'].apply(removespace)
```

```python
movies['tags']=movies['overview']+movies['original_language']+movies['genres']+movies['keywords']
```

## importing nltk package

```python
import nltk
from nltk.stem import PorterStemmer
```

```python
ps=PorterStemmer()
```

```python
def stems(text):
    l=[]
    for i in text.split():
        l.append(i)
    return " ".join(l)
```

```python
mdf['tags'] = mdf['tags'].apply(stems)
```

## Measuring Cosine Similarity

```python
from sklearn.metrics.pairwise import cosine_similarity
from scipy import sparse
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
vectorizer = TfidfVectorizer()
tags_tfidf = vectorizer.fit_transform(mdf['tags'])
```

```python
from sklearn.neighbors import NearestNeighbors
```

```python
nn = NearestNeighbors(n_neighbors=10, algorithm='auto', metric='cosine')
nn.fit(tags_tfidf)
```

## Recommending using Similarity

```python
mdf[mdf['title']=='Spider-Man'].index[0]
```

```python
def recommend(movie):
    idx = mdf.index[mdf['title'] == movie].tolist()[0]
    movie_tags_tfidf = vectorizer.transform([mdf.loc[idx,'tags']])
    distances, indices = nn.kneighbors(movie_tags_tfidf, n_neighbors=31)
    recommended_indices = indices.flatten()[1:]
    recommended_movies = mdf.loc[mdf.index.isin(recommended_indices), 'title'].tolist()
    for i,mov in enumerate(recommended_movies):
        print(mov)
```

```python
recommend('Interstellar')
```
